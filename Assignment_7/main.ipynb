```python
def f1(x):
    return 3*x + np.sin(x) - np.exp(x)
def der_f1(x):
    return 3 + np.cos(x) - np.exp(x)
```

```python
a , b = -1.5 , 1.5
epsilon = 1e-6
bisection_iter = []
bisection_x_values = []
reguli_iter = []
reguli_x_values = []
newton_raph_iter = []
newton_raph_values = []
# Bisection method 
iter1 = 0
is_stop = False
if f1(a)*f1(b) < 0:
    while not is_stop:
        if abs(b-a) < epsilon :
            if abs(f1(a)) < abs(f1(b)):
                print(f"the approximate solution of the function is for bisection method {a}")
                is_stop = True 
            else :
                print(f"the approximate solution of the function is for bisection method {b}")
                is_stop = True 
        c = (a + b ) / 2.0
        f1_c = f1(c)
        f1_b = f1(b)
        f1_a = f1(a)
        iter1 += 1
        bisection_iter.append(iter1)
        bisection_x_values.append(c) 
        if f1_a * f1_c < 0.0:
            
            b = c 
        elif f1_b * f1_c < 0.0:
            a = c 
```
