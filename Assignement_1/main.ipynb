## Question-2 

```python
# LCG event generation 
def LCG(n_event,a = 1103515245 , c = 12345 , m = 32768 , normalise = True ):
    event_list = []
    x_i = 15 
    for i in range(n_event):
        y_i = x_i/m
        event_list.append(y_i)
        x_i = (a * x_i + c )% m
    return event_list
```

```python
lcg_random_events = LCG(1000)
x_values = lcg_random_events
y_values = lcg_random_events[5:] + lcg_random_events [ :5]
plt.figure(120)
plt.scatter(x_values , y_values)
plt.show()
```

```python
# measure pi-value for n = 1000 random number 
n = 1000
x_points = LCG(1000)
y_points = x_points[7:] + x_points[:7]
points_inside_circle = []
x_red_points = []
y_red_points = []
x_blue_points = []
y_blue_points = []
n_in_circle = 0
for i in range(n):
    r = x_points[i]**2 + y_points[i]**2 
    if r <= 1 :
        x_red_points.append(x_points[i])
        y_red_points.append(y_points[i])
        n_in_circle += 1 
    else :
        x_blue_points.append(x_points[i])
        y_blue_points.append(y_points[i])
plt.figure()
plt.scatter(x_blue_points,y_blue_points)
plt.scatter(x_red_points, y_red_points)
pi_value = 4 * n_in_circle / n
print(pi_value)
```

```python
n_event_list = list(range(20,2000,20))
pi_values_list = []
for n in n_event_list:
    x = LCG(n)
    y = x[10 : ] + x[ : 10]
    n_in_circle = 0
    for i in range(n):
        r = x[i]**2 + y[i]**2 
        if r <= 1 :
            n_in_circle += 1 
        else :
            pass
    pi_value  = 4.0 * n_in_circle / n 
    pi_values_list.append(pi_value)

pi_average_value = sum(pi_values_list[int(len(pi_values_list)/2.0) :])/len(pi_values_list[int(len(pi_values_list)/2.0) :])
print(f"Average value of pi is {pi_average_value}")
plt.figure()
plt.plot(n_event_list,pi_values_list)
plt.xlabel("nevents")
plt.ylabel("pi-values")
plt.show()
```
